```r
library(gplots)
paste0 <- function( ..., sep="" ) paste( ..., sep = sep )
holder="/tmp/tmp.C9Dw9Ebfaw"
```

```r
#create breaks and colors
my.breaks=seq(0,0.8,length.out=101)       #101 breaks
my.colors<-colorpanel(100,rgb(49,105,154,25,maxColorValue=255),
rgb(255,255,255,25,maxColorValue=255),rgb(164,25,32,25,maxColorValue=255))       #100 colors

#load and peek at data
e<-read.table(paste0(holder,"_heatmap"))
e
```

```r
#do for ward.D and ward.D2
for(curr_method in c("ward.D","ward.D2"))
{
#  pdf(paste0("/home/mzabidi/TNBC_samples_",curr_method,".pdf"))

    #load function
    hclust2 <- function(x, method=curr_method, ...)
        hclust(x, method=method, ...)
    
    title=paste0("TNBC sample signatures\nmethod = ",curr_method)
    v=heatmap.2(as.matrix(t(e)),main=title,dendrogram='both',trace="none",cexCol=0.5,cexRow=0.8,
              density.info="none",key=TRUE,symkey=TRUE,breaks=my.breaks,scale="none",col=my.colors,
              hclustfun=hclust2)

#  dev.off()
    
    #write sample arrangement to file
    fileConn<-file(paste0(holder,"_order_",curr_method))
    writeLines(c(row.names(v$carpet)), fileConn)
    close(fileConn)

}
```
